### 5 . Perform meaningful visualizations using the dataset (min 3)


```python
# plt.style.available to see style options available
```

```python
plt.style.use('seaborn-darkgrid')
sns.displot(df.Classification)
plt.title("Histogram for classification",fontsize =14)
plt.xticks(fontsize = 12)
```

         
      The above histogram clearly shows that ,there are three species in the classification and all are equally distributed, 50 records for each classification.Now lets analyse how each measurements varies among the 3 classification.  For that we will use a grouped boxplot with classification as x axis as  and measurements as y axis.Here we  have 4 different measurements for each record. So we will use a for loop.

```python
cols = df.columns
ncols = (cols[:-1]).tolist() #Convert pandas.core.indexes.base.Index type to list for further processing
```

```python
plt.style.use('seaborn-pastel')
plt.figure(figsize = (12,8)) # set fig size for the plots.
for i in ncols:
    plt.subplot(2,2,ncols.index(i)+1)
    sns.boxplot(data =df, x = 'Classification', y = i)
    plt.title("{} for each class".format(i),fontsize =13)
    plt.xlabel("") # xlabel is deleted ,instead each category will be displayed 
    plt.xticks(fontsize = 12)
    
```

### Conclusions.
* Iris setosa has small petal with wide sepals.
* Iris-versicolor has medium size petal with medium size sepals.
* Iris-virginica has large petals with lengthy sepals.

We will do further detailed analysis on each species,first Iris setosa.

### Detailed study on Iris-setosa

```python
dfIris = df[df['Classification']=='Iris-setosa']
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(dfIris.corr(),annot =True)
plt.title("Heatmap for Iris-setosa measurements")
```

We see that there is only a positive corelation between sepal length and sepal width. Other measurements doesn't have a significant corelation.
