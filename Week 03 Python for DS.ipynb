### 4.4. Find the minimum and maximum value of insulin level in the data.

```python
print("Maximun insulin level in the data set is {}".format(df.Insulin.max()))
print("Minimum insulin level in the data set is {}".format(df.Insulin.min()))
```

### 4.5. Create a histogram of "Age".

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(df.Age)
plt.title("diabetes_model Age histogram")
```

### 4.6. Create a scatter plot of "BMI" vs "Glucose".

```python
x,y = 'BMI','Glucose'
sns.scatterplot(data = df, x = x, y = y)
plt.title("diabetes_model {} vs {}".format(x,y))
```

### 4.7. Replace the min value of "BMI" with mean value of BMI

```python
mean = round(df.BMI.mean(),2) # mean value has many decimal places.So store its rounded value to mean variable
mean
```

```python
min = df.BMI.min() # Assign min value to a variable to avoid multiple min calls during replace command
min
```

```python
df['BMI'].replace({min:mean},inplace =True)
print("min : {} , mean{} ".format(df.BMI.min(),round(df.BMI.mean(),2)))
```

Replaced min BMI from zero to mean value(32.21). Now the min BMI is 18.2 and tne mean is 32.58.

### 4.8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
sns.boxplot(data = df, x = 'Outcome', y = 'Glucose')
```
