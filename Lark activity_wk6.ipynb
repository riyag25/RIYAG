## 2.Find out how many countries with respect to the region are available in the dataset

```python
regions = df.groupby(by=['Region'])['country'].nunique().sort_values()
regions
```

```python
regions.plot(kind ='bar')
plt.title("Number of countries in Region")
plt.yticks(np.arange(0,20,2))
plt.show()
```

```python
# We will first group the countries by region.Then get the mean GDP for each region.
```

```python
df.groupby(by='Region')['GDP per capita (current US$)'].mean().sort_values().plot(kind ='bar')
plt.title("Mean GDP per capita (current US$)")
plt.show()
```

***
Northern America has the highest GDP per capita (current US$) on average.

***

```python
df.groupby(by='Region')['International trade: Imports (million US$)'].mean().sort_values().plot(kind ='bar')
plt.title("Mean International trade: Imports (million US$) ")
plt.show()
```

***
Northern America has the highest International trade: Imports (million US$) on average
***

## 4.Find out which region is more consistent with respect to GDP per capita (current US$ )

```python
#To check consistence, check the variation. We will use std.
df.groupby(by='Region')['GDP per capita (current US$)'].std().sort_values().plot(kind ='bar')
plt.title("std of GDP per capita (current US$)")
plt.show()
```

***
As per the above graph, WesternAfrica has the least variation(in terms of standard deviation) in GDP per capita. WesternAfrica is also the country with lowest mean GDP per capita.
But among the top 3 regions with high mean GDP per capita, Oceania is more consistent.
