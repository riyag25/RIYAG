```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.Read the data into python

```python
df = pd.read_csv('week10.csv')
df.head()
```

## 2. Check for null values

```python
df.isna().sum()
```

## 3.Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations.

```python
select = ['Hardness','Sulfate', 'Organic_carbon']
sns.pairplot(df[select])
```

```python
df[select].plot(kind = 'box')
```

As per given data,  the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ are normally distributed with outliers.From the pair plot, we can observe that there is no significant corelation between the 3 features.

## 4. Carry out correlation analysis on the dataset and find the correlation factor between
* a. ph and Chloramines
* b. ph and Conductivity
## What do we infer from this correlation value

```python
corr = df.corr()
sns.heatmap(corr, annot =True)
```

There is no highly corelated features in the dataset.

### a. ph and Chloramines

```python
df[['ph','Chloramines']].corr()
```

```python
sns.scatterplot(data =df, x = 'Chloramines', y = 'ph')
plt.title('Chloramines vs ph')
plt.show()
```

Corelation factor is -.024768,which is approximately equal to zero.We can conclude that,there is no relevant corelation between ph and Chloramines.

### b. ph and Conductivity

```python
df[['ph','Conductivity']].corr()
```

```python
sns.scatterplot(data =df, x = 'Conductivity', y = 'ph')
plt.title('Conductivity vs ph')
plt.show()
```

Corelation factor is .014128 ,which is approximately equal to zero.We can conclude that,there is no relevant corelation between ph and Conductivity.
