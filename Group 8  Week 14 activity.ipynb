```python
aid_df = aid_df.drop([20,  81, 87, 88, 101, 102])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
aid_df['country'] = le.fit_transform(aid_df['country'])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
n =np.arange(2,8)
for i in n:
    model = KMeans(n_clusters = i, init ='k-means++', random_state = 42)
    model.fit(aid_df)
    y = model.fit_predict(aid_df)
    wcss.append(model.inertia_)
    print( print('n: {}, wcss: {} , Silhouette Score:  {}'.format(i, wcss[i-3],silhouette_score(aid_df, y))))
plt.plot(range(2,8),wcss,'-o')
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()   
    
```

```python
#n = 3
model = KMeans(n_clusters = 3, init ='k-means++', random_state = 42)
y = model.fit_predict(aid_df)
```

```python
aid_df['category'] = y
```

```python

sns.catplot(data=aid_df,orient ='h' , palette="Set2", col = 'category',)
```

```python
aid_df.groupby('category').mean()
```

As per the above data , category 0 countries has poor indicators.So they are in direct need of  aid.

```python
Aid_needed = aid_df[aid_df['category']==0]
Aid_needed['country'] =le.inverse_transform(Aid_needed['country'])
```
